# Distribution Analysis: Categorical Variables

```python
# Get categorical and numerical columns from client data
all_categorical = clients.select_dtypes(include=['object']).columns.tolist()
all_numerical = clients.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Exclude ID and date columns from categorical
exclude_patterns = ['id', 'date_']
categorical_cols = [col for col in all_categorical 
                     if not any(pattern in col.lower() for pattern in exclude_patterns)]

# Exclude target variable 'churn' from numerical visualization
numerical_cols = [col for col in all_numerical if col != 'churn']

print("Categorical columns (excluding ID and dates):", categorical_cols)
print("\nNumerical columns (excluding target 'churn'):", numerical_cols)
```

```python
# Visualize CATEGORICAL variables
fig, axes = plt.subplots(len(categorical_cols), 1, figsize=(12, 4 * len(categorical_cols)))

# Ensure axes is iterable (if only 1 categorical, axes is 1D)
if len(categorical_cols) == 1:
    axes = [axes]

for idx, col in enumerate(categorical_cols):
    value_counts = clients[col].value_counts()
    axes[idx].bar(range(len(value_counts)), value_counts.values, color='steelblue', edgecolor='black')
    axes[idx].set_xticks(range(len(value_counts)))
    axes[idx].set_xticklabels(value_counts.index, rotation=45, ha='right')
    axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
    axes[idx].set_ylabel('Frequency', fontsize=10)
    axes[idx].grid(axis='y', alpha=0.3)
    
    # Add value labels on bars
    for i, v in enumerate(value_counts.values):
        axes[idx].text(i, v + 5, str(v), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

# Distribution Analysis: Numerical Variables

```python
# Visualize NUMERICAL variables - Histograms only
num_cols_per_row = 3
num_rows = (len(numerical_cols) + num_cols_per_row - 1) // num_cols_per_row

fig, axes = plt.subplots(num_rows, num_cols_per_row, figsize=(16, 4 * num_rows))
axes = axes.flatten()  

for idx, col in enumerate(numerical_cols):
    axes[idx].hist(clients[col].dropna(), bins=30, color='steelblue', alpha=0.7, edgecolor='black')
    
    axes[idx].set_title(f'Distribution of {col}', fontsize=11, fontweight='bold')
    axes[idx].set_xlabel(col, fontsize=10)
    axes[idx].set_ylabel('Frequency', fontsize=10)
    axes[idx].grid(alpha=0.3)

# Remove extra subplots
for idx in range(len(numerical_cols), len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.show()
```

```python
# Visualize NUMERICAL variables - Box Plots (for outlier detection)
num_cols_per_row = 3
num_rows = (len(numerical_cols) + num_cols_per_row - 1) // num_cols_per_row

fig, axes = plt.subplots(num_rows, num_cols_per_row, figsize=(16, 4 * num_rows))
axes = axes.flatten()

for idx, col in enumerate(numerical_cols):
    axes[idx].boxplot(clients[col].dropna(), vert=True, patch_artist=True,
                      boxprops=dict(facecolor='lightblue', alpha=0.7),
                      medianprops=dict(color='red', linewidth=2),
                      whiskerprops=dict(color='black'),
                      capprops=dict(color='black'))
    axes[idx].set_title(f'Box Plot: {col}', fontsize=11, fontweight='bold')
    axes[idx].set_ylabel(col, fontsize=10)
    axes[idx].grid(axis='y', alpha=0.3)
    
    # Add statistics
    q1 = clients[col].quantile(0.25)
    median = clients[col].median()
    q3 = clients[col].quantile(0.75)
    axes[idx].text(1.15, median, f'Median: {median:.2f}', fontsize=9, va='center')

# Remove extra subplots
for idx in range(len(numerical_cols), len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.show()
```

# Target Variable Analysis: Churn Distribution

```python
# Visualize CHURN distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Bar plot with counts
churn_counts = clients['churn'].value_counts()
colors = ['#2ecc71', '#e74c3c']  # Green for 0 (stayed), Red for 1 (churned)
axes[0].bar(churn_counts.index, churn_counts.values, color=colors, edgecolor='black', width=0.6)
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['No Churn (0)', 'Churned (1)'], fontsize=11)
axes[0].set_title('Churn Distribution - Count', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Number of Customers', fontsize=10)
axes[0].grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, v in enumerate(churn_counts.values):
    axes[0].text(i, v + 20, str(v), ha='center', va='bottom', fontsize=11, fontweight='bold')

# Plot 2: Pie chart with percentages
churn_pct = clients['churn'].value_counts(normalize=True) * 100
labels = [f'No Churn\n({churn_pct[0]:.1f}%)', f'Churned\n({churn_pct[1]:.1f}%)']
axes[1].pie(churn_counts.values, labels=labels, colors=colors, autopct='', startangle=90,
            textprops={'fontsize': 11, 'fontweight': 'bold'}, wedgeprops={'edgecolor': 'black'})
axes[1].set_title('Churn Distribution - Percentage', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Bivariate Analysis vs Churn

```python
# Prepare a working copy and derive contract duration (proxy for contract profile)
bi_data = clients.copy()

for dcol in ['date_activ', 'date_end']:
    if dcol in bi_data.columns:
        bi_data[dcol] = pd.to_datetime(bi_data[dcol], errors='coerce')

if {'date_activ', 'date_end'}.issubset(bi_data.columns):
    bi_data['contract_duration_days'] = (bi_data['date_end'] - bi_data['date_activ']).dt.days


def plot_categorical_churn_stacked(df, col, top_n=10):
    temp = df[[col, 'churn']].dropna().copy()
    if temp.empty:
        print(f"No data available for {col}")
        return

    # Keep top categories to avoid overloaded charts
    top_categories = temp[col].value_counts().head(top_n).index
    temp = temp[temp[col].isin(top_categories)]

    stacked = temp.groupby([col, 'churn']).size().unstack(fill_value=0)

    # Ensure both churn classes exist for consistent colors
    for klass in [0, 1]:
        if klass not in stacked.columns:
            stacked[klass] = 0
    stacked = stacked[[0, 1]]

    # Sort by churn rate descending
    churn_rate = stacked[1] / (stacked[0] + stacked[1])
    stacked = stacked.loc[churn_rate.sort_values(ascending=False).index]

    fig, ax = plt.subplots(figsize=(12, 5))
    stacked.plot(
        kind='bar',
        stacked=True,
        color=['#1f77b4', '#ff7f0e'],  # Blue / Orange
        edgecolor='white',
        width=0.75,
        ax=ax
    )

    ax.set_title(f'{col} vs Churn (stacked count)', fontsize=12, fontweight='bold')
    ax.set_xlabel(col)
    ax.set_ylabel('Number of customers')
    ax.tick_params(axis='x', rotation=45)
    ax.grid(axis='y', alpha=0.2)
    ax.legend(['No churn (0)', 'Churn (1)'], title='Class', frameon=False)

    plt.tight_layout()
    plt.show()


# Requested categorical views
for c in ['channel_sales', 'origin_up', 'nb_prod_act', 'has_gas']:
    if c in bi_data.columns:
        plot_categorical_churn_stacked(bi_data, c, top_n=10)
    else:
        print(f"Column not found: {c}")
```

```python
# Numerical bivariate analysis: churn rate by bins + boxplots by churn
num_vars_to_analyze = [
    'num_years_antig',
    'net_margin',
    'margin_net_pow_ele',
    'margin_gross_pow_ele',
    'contract_duration_days',
    'pow_max',
    'imp_cons',
    'forecast_meter_rent_12m',
    'forecast_cons_12m'
]
num_vars_to_analyze = [c for c in num_vars_to_analyze if c in bi_data.columns]

for col in num_vars_to_analyze:
    temp = bi_data[[col, 'churn']].dropna().copy()
    if temp.empty:
        print(f"No data for {col}")
        continue

    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    # Left: boxplot to compare distributions by churn class
    temp.boxplot(column=col, by='churn', ax=axes[0], grid=False)
    axes[0].set_title(f'{col} by churn class')
    axes[0].set_xlabel('churn (0=No, 1=Yes)')
    axes[0].set_ylabel(col)

    # Right: churn rate across quantile bins (behavioral trend)
    temp['bin'] = pd.qcut(temp[col], q=5, duplicates='drop')
    churn_by_bin = temp.groupby('bin', observed=False)['churn'].mean().mul(100)
    churn_by_bin.plot(kind='bar', ax=axes[1], color='steelblue', edgecolor='black')
    axes[1].set_title(f'Churn rate across {col} bins')
    axes[1].set_ylabel('Churn rate (%)')
    axes[1].set_xlabel(col)
    axes[1].tick_params(axis='x', rotation=35)
    axes[1].grid(axis='y', alpha=0.2)

    plt.suptitle('')
    plt.tight_layout()
    plt.show()
```
